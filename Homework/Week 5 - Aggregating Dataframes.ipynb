```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('nyc_weather_2018.csv')
data2 = pd.read_csv('weather_stations.csv')
data1.head()
```

```python
data1.shape
```

```python
data2.shape
```

```python
data = pd.merge(data1,data2,left_on='station', right_on='id', how='left')
data = data.drop('id',1)
data = data.drop_duplicates(subset=['station','datatype'])
data.head()
```

```python
data.shape
```

```python
snow = data.query('datatype == "SNOW" and name.str.contains("NJ US")', engine='python')
snow.elevation.mean()
```
