```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('earthquakes.csv', sep=',')
data.head()
```

# How many earthquake have magType 'ml' and type 'explosion'?

```python
len(data[(data['type']=='explosion')&(data['magType']=='ml')])
```

# Calculate the mean of mag in Alaska and the distance is smaller than 100km

```python
Alaska = data[data['place'].str.contains('Alaska')]
d = Alaska['place'].str.findall('\d+')
km = [int(x[0]) if x else 0 for x in d] 
Alaska['km'] = km
```

```python
Alaska.head()
```

```python
Alaska[Alaska['km']<100].mag.mean()
```

# Calculate the mean of mag in each quarter

```python
da = pd.to_datetime(data['time'], unit='ms')
q = pd.DatetimeIndex(da).quarter
data['quarter'] = q
data.head()
```

```python
pd.pivot_table(data, values='mag', index=['quarter'],
                    aggfunc=np.mean)
```
